```python
import os
import numpy as np
project_dir = os.path.join('D:/', 'UG', 'Research', 'Dream_Neural_Decoding',
                           'Shannon_Big_Big_Dream_Project','project_directory')
THINGS_dir = os.path.join(project_dir, 'eeg_dataset', 'wake_data', 'THINGS')
```

#### I use the subject 1 as my training data

```python
# Load THINGS training data  
def load_train(project_dir,sub):
    THINGS_train_dir = os.path.join(THINGS_dir, 
                                    'preprocessed_data','sub-'+format(sub,'02'),
                                    'preprocessed_eeg_training.npy')
    THINGS_train_data = np.load(THINGS_train_dir, allow_pickle=True).item()
    train_data = THINGS_train_data['preprocessed_eeg_data']
    
    train_reps = train_data.shape[1]
    train_channels = train_data.shape[2]
    train_times = train_data.shape[3]
    
    X_n = train_data[0:10,:,:,:].reshape((10*train_reps, train_channels, train_times))
    X_p = train_data[10:20,:,:,:].reshape((10*train_reps, train_channels, train_times))
    X = np.concatenate((X_n,X_p))

    y_n = [0]*10*train_reps
    y_p = [1]*10*train_reps
    y = np.concatenate((y_n,y_p))
    
    return X,y
```

```python
X_1,y_1 = load_train(project_dir,1)
X_2,y_2 = load_train(project_dir,2)

X = np.concatenate((X_1,X_2))
y = np.concatenate((y_1,y_2))
```

```python
from matplotlib import pyplot as plt
from tqdm import tqdm
from sklearn.linear_model import LogisticRegression

for t in tqdm(range(X.shape[2]), desc='decoding training'):
    logreg = LogisticRegression().fit(X[:,:,t], y)
```

```python
from sklearn.model_selection import learning_curve
from sklearn.model_selection import LearningCurveDisplay

common_params = {
"X": X[:,:,25],
"y": y,
"train_sizes": np.linspace(0.1, 1.0, 5),
"cv": 5,
"score_type": "both",
"n_jobs": -1,
"line_kw": {"marker": "o"},
"std_display_style": "fill_between",
"score_name": "Accuracy",}
LearningCurveDisplay.from_estimator(logreg, **common_params)
```

#### And use subject 2 as my test data

```python
# Load THINGS test data  
THINGS_test_dir = os.path.join(THINGS_dir, 
                                'preprocessed_data','sub-'+format(2,'02'),
                                'preprocessed_eeg_training.npy')
THINGS_test_data = np.load(THINGS_test_dir, allow_pickle=True).item()
test_data = THINGS_test_data['preprocessed_eeg_data']
print('shape of test data:', test_data.shape)
```

```python
test_reps = test_data.shape[1]
test_channels = test_data.shape[2]
test_times = test_data.shape[3]
```

```python
X_0_test = test_data[0:10,:,:,:].reshape((10*test_reps, test_channels, test_times))
X_1_test = test_data[10:20,:,:,:].reshape((10*test_reps, test_channels, test_times))
X_test = np.concatenate((X_0_test,X_1_test))

y_0_test = [0]*10*test_reps
y_1_test = [1]*10*test_reps
y_test = np.concatenate((y_0_test,y_1_test))
```

```python
test_conds = X_test.shape[0]
result = np.empty((test_conds, test_times))

plt.figure()
for cond in range(test_conds): 
    for T in range(test_times):
        result[cond][T] = logreg.predict_proba(X_test[cond,:,T].reshape(1,-1))[0][0]
    plt.plot(range(test_times), result[cond], 
             # color=color, 
             alpha=0.1)
plt.xlabel('Time (10ms)')
plt.ylabel('Probability')
plt.title('Image decoder')
plt.show()
```
